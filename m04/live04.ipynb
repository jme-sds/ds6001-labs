## Congress API

```python
dotenv.load_dotenv()
congress = os.getenv('congress')
```

```python
url = 'https://httpbin.org/user-agent'
r = requests.get(url)
mjson = r.json()
useragent = mjson['user-agent']
```

```python
headers = {'User-Agent': useragent,
           'From':'jme3qd@virginia.edu'}
```

Four things to look for in API docs
1. root
2. endpoint for data source you need
3. parameters
4. headers, or other instructions for supplying your API

```python
root = 'https://api.congress.gov/v3'
endpoint = '/member'
params = {'format':'json',
          'limit':250,
          'offset':'0',
          'currentMember': 'true',
          'api_key': congress}
```

```python
r = requests.get(root+endpoint,params=params,headers=headers)
r
```

```python
myjson = json.loads(r.text)
congressdf = pd.json_normalize(myjson,record_path=['members'])
congressdf
```
